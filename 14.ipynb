```python
import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42
)

def model_fn(a_layer=None):
    model = keras.Sequential(name='fashion')
    model.add(keras.layers.Input(shape=(28,28)))
    model.add(keras.layers.Flatten(name='flatten'))
    model.add(keras.layers.Dense(100, activation='relu', name='hidden'))
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dense(10, activation='softmax', name='output'))
    return model

# model = model_fn()
model = model_fn(keras.layers.Dropout(0.3))     # 30% drop out
model.summary()

adam = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=adam,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_scaled, train_target,
                    epochs=5,
                    validation_data=(val_scaled, val_target),
                    verbose=0)

print(history.history.keys())

import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.xlabel('epoch')
plt.legend()
plt.show()

```

 Dropout : 훈련과정에서 층에 있는 일부 뉴런을 꺼서 (출력을 0으로) 과대적합을 해소

 - 결과가 특정 뉴런에 과대하게 의존하는 것을 방지
 - 기본적으로 Hidden층 뒤에 추가한다.

```python
import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42
)

def model_fn(a_layer=None):
    model = keras.Sequential(name='fashion')
    model.add(keras.layers.Input(shape=(28,28)))
    model.add(keras.layers.Flatten(name='flatten'))
    model.add(keras.layers.Dense(100, activation='relu', name='hidden'))
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dense(10, activation='softmax', name='output'))
    return model

# model = model_fn()
model = model_fn(keras.layers.Dropout(0.3))     # 30% drop out
model.summary()

adam = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=adam,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_scaled, train_target,
                    epochs=11,
                    validation_data=(val_scaled, val_target))

model.save('model-fashion.keras')                          # 모델 구조와 파라메타 저장
model.save_weights('model-fashion.weights.h5')             # 파라메타만 저장
```

```python
import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42
)

def model_fn(a_layer=None):
    model = keras.Sequential(name='fashion')
    model.add(keras.layers.Input(shape=(28,28)))
    model.add(keras.layers.Flatten(name='flatten'))
    model.add(keras.layers.Dense(100, activation='relu', name='hidden'))
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dense(10, activation='softmax', name='output'))
    return model

# model = model_fn()
model = model_fn(keras.layers.Dropout(0.3))     # 30% drop out
model.load_weights('model-fashion.weights.h5')

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.evaluate(val_scaled, val_target)

# 저장 모델 사용
# model = model_fn(keras.layers.Dropout(0.3))
model=keras.models.load_model('model-fashion.keras')
model.evaluate(val_scaled, val_target)
```

Callback: 훈련 중간에 작업을 수행

```python
model = model_fn(keras.layers.Dropout(0.3))

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-fashion-model.keras', save_best_only=True)
model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target),callbacks=[checkpoint_cb])

model.evaluate(val_scaled, val_target)

model = keras.models.load_model('best-fashion-model.keras')
model.evaluate(val_scaled, val_target)
```

save_best_only=True는 가장 낮은 검증 손실 모델을 저장

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])

checkpoint_cb = keras.callbacks.ModelCheckpoint('best-fashion-model.keras', save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target),
callbacks=[checkpoint_cb, early_stopping_cb])

model.evaluate(val_scaled, val_target)

print(early_stopping_cb.stopped_epoch)

model = keras.models.load_model('best-fashion-model.keras')
model.evaluate(val_scaled, val_target)
```

- patience=2 2번 연속검증 점수가 향상되지 않으면 중지
- restore_best_weights = True 가장 최선의 파라메타로 되돌림
