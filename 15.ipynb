```python
import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0
test_scaled = test_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)

def model_fn():
  model = keras.Sequential(name='fashion')
  model.add(keras.layers.Input(shape=(28,28)))
  model.add(keras.layers.Flatten(name='flatten'))
  model.add(keras.layers.Dense(100, activation='relu', name='hiden'))
  model.add(keras.layers.Dense(10, activation='softmax', name='output'))
  return model

model = model_fn()
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])

checkpoint_cb = keras.callbacks.ModelCheckpoint('best-fashion-model.keras', save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

model.fit(train_scaled, train_target, epochs=20, verbose=0, validation_data=(val_scaled, val_target),
callbacks=[checkpoint_cb, early_stopping_cb])

model.evaluate(val_scaled, val_target)

print(early_stopping_cb.stopped_epoch)

model = keras.models.load_model('best-fashion-model.keras')
model.evaluate(val_scaled, val_target)

History=model.fit(train_scaled, train_target, epochs=4, verbose=1, validation_data=(val_scaled, val_target))

import numpy as np
print(np.round(model.predict(val_scaled[3:4]),3))

output_weights = model.get_layer('output').get_weights()[0]

first_neuron_weights = output_weights[:,0]
print(first_neuron_weights[0:5])
```
